```python
# set pdf input folder and json output folders
pdf_folder  = '/content/drive/MyDrive/pdf_output/'
json_output = '/content/drive/MyDrive/json_output/'
```

```python
# create output folder if it does not exist

if os.path.isdir(json_output) == False:
  os.mkdir(json_output)
```

```python
import camelot
import os
```

```python
# generate list of pdfs to scrape
files = []
for file_name in os.listdir(pdf_folder):
  if file_name.endswith('pdf'):
    files.append(file_name)
```

```python
# example of output in a pandas dataframe

table_example = camelot.read_pdf('/content/drive/MyDrive/pdf_output/CHIPPEWAS_OF_RAMA_FIRST_NATION__ON_2019-20.pdf', flavor='stream',pages='all')
table_example[8].df
```

```python
# cycle through all PDFs in specified folder and export all detected tables as JSON files to specified output folder

for pdfname in files:
  pdf_path = pdf_folder + pdfname
  output_path = json_output + pdfname + '/'
  if os.path.isdir(output_path) == False:
    os.mkdir(output_path)
  tables = camelot.read_pdf(pdf_path, flavor='stream', pages='all')
  for i in range(0,len(tables)):
    tables[i].to_json(f'{output_path}{pdfname}_table{i+1}.json')
```
